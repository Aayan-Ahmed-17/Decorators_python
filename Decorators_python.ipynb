> **Decorators**

def:
func | takes func | return enhanced func



```python
"""
Let's see an example in which a simple Python decorator adds behavior before and after a function call.
"""

def decorator(func):

  def wrapper():
    print("Before calling function")
    func()
    print("After calling function")

  return wrapper

@decorator
def greet():
  print("Hello")

greet()
```

**Decorator with args and kwargs**

```python
def decorator_with_arg(func):
  def wrapper(*args, **kwargs):
    print("args=>", args)
    result = func(*args, **kwargs)
    print("kwargs=>", kwargs)

    return result

  return wrapper

@decorator_with_arg
def add(a, b):
  print(a + b)

add(2, 4)
```

**Practical Use case of decorators with args and kwargs**

```python
"""
Real-world context: create a decorator to debug production code or monitor which functions are slow.
"""
import time
from datetime import datetime

def log_calls(func):
  def wrapper(*args, **kwargs):
    print(f"{datetime.now()} Calling Function {func.__name__} with *args={args} & **kwargs={kwargs}")
    result = func(*args, **kwargs)
    print(f"{datetime.now()} Function named {func.__name__} returned {result}")
    return result

  return wrapper

@log_calls
def add(a, b):
  print(a + b)

add(2, 3)
```

```python
"""
Use Case 2: Timing Functions (Performance Monitoring)
"""

import time

def timing_decorator(func):
  def wrapper(*args, **kwargs):
    start = time.time()
    result = func(*args, **kwargs)
    end = time.time()
    difference = end - start
    print(f"{func.__name__} with args: {args}, kwargs: {kwargs} took {difference:.4f}s")
    return result

  return wrapper

@timing_decorator
def slow_function(a, b, c):
    time.sleep(2)
    return "Done!"

@timing_decorator
def fetch_data_from_api(a, b, c, **kwargs):
    # Simulating API call
    time.sleep(0.5)
    return {"data": "results"}

slow_function(2, "adljslf", True)
fetch_data_from_api(False, [2, 3, 4], "hello", extra = "done")
```

> **Decorators**

def:
func | takes func | return enhanced func



```python
"""
Let's see an example in which a simple Python decorator adds behavior before and after a function call.
"""

def decorator(func):

  def wrapper():
    print("Before calling function")
    func()
    print("After calling function")

  return wrapper

@decorator
def greet():
  print("Hello")

greet()
```

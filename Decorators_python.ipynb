> **Decorators**

def:
func | takes func | return enhanced func



```python
"""
Let's see an example in which a simple Python decorator adds behavior before and after a function call.
"""

def decorator(func):

  def wrapper():
    print("Before calling function")
    func()
    print("After calling function")

  return wrapper

@decorator
def greet():
  print("Hello")

greet()
```

**Decorator with args and kwargs**

```python
def decorator_with_arg(func):
  def wrapper(*args, **kwargs):
    print("args=>", args)
    result = func(*args, **kwargs)
    print("kwargs=>", kwargs)

    return result

  return wrapper

@decorator_with_arg
def add(a, b):
  print(a + b)

add(2, 4)
```
